```python
sodh = pd.read_parquet("../data/SDoH.parquet")
sodh.GEOID = pd.to_numeric(sodh.GEOID, errors="coerce")
```

```python
sodh.GEOID.isna().sum()/len(sodh)
```

```python
sodh_by_tract = sodh.groupby("GEOID").mean().reset_index()
sodh_by_tract.GEOID = sodh_by_tract.GEOID.round(0).astype("Int64").astype("string")
sodh_by_tract
```
